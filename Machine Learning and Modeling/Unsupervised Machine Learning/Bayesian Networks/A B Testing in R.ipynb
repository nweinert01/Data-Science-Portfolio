# A/B Testing in R

# Bayesian Networks

```r
trials <- 10000
```

```r
alpha <- 8
beta <- 2
```

```r
samplesA <- rbeta(trials, 95+alpha, 22 + beta)
samplesB <- rbeta(trials, 73+alpha, 46 + beta)
```

```r
Bsuperior <- sum(samplesB > samplesA) / trials
```

```r
Bsuperior
```

```r
print(Bsuperior)
```
