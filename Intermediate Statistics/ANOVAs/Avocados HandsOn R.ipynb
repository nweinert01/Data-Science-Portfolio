### Avacado Hands On - R
Does the average price of avocados differ between Albany, Houston, and Seattle?


```r
# Load Pakages
library("dplyr")
library("rcompanion")
library("car")
```

```r
# Load Data
Avocado <- read.csv("avocados.csv")
```

```r
avocadoF <- na.omit(Avocado %>% filter(region %in% c("Albany", "Houston", "Seattle")))
```

```r
# Normality
plotNormalHistogram(avocadoF$AveragePrice)
```

```r
avocadoFSQRT <- sqrt(avocadoF$AveragePrice)
plotNormalHistogram(avocadoFSQRT)
```

```r
avocadoFLOG <- log(avocadoF$AveragePrice)
plotNormalHistogram(avocadoFLOG)
# Log is most normal distribution
```

## Test for Homogeneity of Variance

```r
bartlett.test(avocadoFLOG ~ region, data=avocadoF)
```

### Bartlett's Test interpretation
P Value < .05 means that this test has violated the assumption of homogeneity of variance

## Test for Unequal Variance

```r
ANOVA1 <- lm(avocadoFLOG ~ region, data=avocadoF)
Anova(ANOVA1, Type ="II", white.adjust=TRUE)
```

## Post Hocs with Unequal Variance

```r
pairwise.t.test(avocadoF$AveragePrice, avocadoF$region, p.adjust = "bonferroni", pool.sd=FALSE)
```

## Find Means

```r
avocadoMeans <- avocadoF %>% group_by(region) %>% summarize(Mean=mean(AveragePrice))
print(avocadoMeans)
```
