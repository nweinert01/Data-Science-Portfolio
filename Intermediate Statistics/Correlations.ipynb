# Correlation

### Correlations can be done on two continuous variables, to determine the relationship between them. As a reminder, a correlation can be between zero and one, and either positive or negative. The larger it is, the more closely related the two variables are.

## Import Packages

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

## Import Data

```python
cruise_ship = pd.read_csv(r'C:\Users\nikki\OneDrive\Pictures\Data Science\105 Intermediate Statistics\cruise_ship.csv')
cruise_ship.head(15)
```

### Run One Correlation

```python
# Are the number of passengers and cabins related to one another?
```

```python
cruise_ship['passngrs'].corr(cruise_ship['Cabins'])
```

### RESULTS: It should not come as a surprise to you that the number of passengers and the number of cabins are very related. .97 is about as strong as a correlation gets! This correlation is also positive, which means that as the number of passengers increase, so does the number of cabins, and vice versa.

# Create a Correlation Matrix

```python
# How all the data relates to one another
```

### Drop non-continuous variable

```python
cruise_ship1 = cruise_ship.drop(['Ship', 'Line'], axis=1)
cruise_ship1.head(10)
```

### Use .corr() on entire dataset

```python
cruise_ship1.corr(method='pearson')
```

### Make Correlation Matrix Pretty

```python
cruise_ship1.corr(method='pearson').style.format("{:.2}").background_gradient(cmap=plt.get_cmap('coolwarm'), axis=1)
```

### Use sns.heatmap()

```python
sns.heatmap(cruise_ship1.corr(), annot=True)
```
