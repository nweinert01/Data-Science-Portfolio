# Transformations in R

```r
# Load the CSV file from the same directory
anime <- read.csv("anime.csv")
```

```r
# Show the first 5 rows of the score column
head(anime$score, 5)
```

### Visualizing Transformations

```r
# Install and import rcompanion
library(rcompanion)
```

```r
# Negatively Skewed Data
plotNormalHistogram(anime$score)
```

# Positively Skewed Data

```r
plotNormalHistogram(anime$scored_by)
```

```r
# Using sqrt() for positively skewed data
anime$scored_bySQRT <- sqrt(anime$scored_by)
```

```r
#Rerun the histogram
plotNormalHistogram(anime$scored_bySQRT)
```

```r
# Make a larger transformation
anime$scored_byLOG <- log(anime$scored_by)
```

```r
# Rerun the histogram to check
plotNormalHistogram(anime$scored_byLOG)
```

### Solve the error: Removing Infinite Values

# The transformation made a few of the numbers so small that R gave up, and just gave them the value of "-inf", or infinitely small. Most functions in R will not work with infinite data, including this one. There is, however, a line of code that can save the day.

```r
library("IDPmisc")
anime2 <- NaRV.omit(anime)
```

```r
# Graph it again
plotNormalHistogram(anime2$scored_byLOG)
```

# Transforming Negatively Skewed Data

```r
# The variable aired_from_year is negatively skewed
```

```r
plotNormalHistogram(anime$aired_from_year)
```

### Sqaring the variable

```r
anime$aired_from_yearSQ <- anime$aired_from_year * anime$aired_from_year
```

```r
# Graph it
plotNormalHistogram(anime$aired_from_yearSQ)
```

### Cubing the variable

```r
anime$aired_from_yearCUBE <- anime$aired_from_year ^ 3
```

```r
# Graph it
plotNormalHistogram(anime$aired_from_yearCUBE)
```
